```python
movies_dic={}
```

```python
for i, j in zip(df["Series_Title"], df["Overview"]):
    if isinstance(j, str):  
        movies_dic[i] = nlp(j)  
```

```python
def recommend_movies(movie_name,movies_dic,n=5):
    if movie_name in movies_dic:
        temp={}
        overview=movies_dic[movie_name]
        for i in movies_dic:
            if i!= movie_name:
                temp[i]=overview.similarity(movies_dic[i])
        res=sorted(temp.items(),key=lambda a:a[1],reverse=True)
        return list(dict(res[:n]).keys())
    else:
        return "Movie not found"
```

```python
recommend_movies('From Here to Eternity',movies_dic)
```

```python
recommend_movies("A Hard Day's Night",movies_dic,10)
```
